```python
#Calcul des valeurs propres
print(pca.explained_variance_)
```

```python
# tableau des valeurs propres et variances pour chaque dimension
eig = pd.DataFrame(
    {
        "Dimension" : ["Dim" + str(x + 1) for x in range(5)], 
        "Valeur propre" : pca.explained_variance_,
        "% variance expliquée" : np.round(pca.explained_variance_ratio_ * 100),
        "% cum. var. expliquée" : np.round(np.cumsum(pca.explained_variance_ratio_) * 100)
    },
    columns = ["Dimension", "Valeur propre", "% variance expliquée", "% cum. var. expliquée"]
)
eig
```

```python
#graphes des variances expliquées par dimension
plt.figure(figsize=(13,10))
g_eig = sns.barplot(x = "Dimension", 
                        y = "% variance expliquée",
                        
                        data = eig)
plt.text(3.25, 26, "20%")
plt.axhline(y = 20, linewidth = .5, linestyle = "--") # 20 = 100 / 5 (nb dimensions)
g_eig.set(ylabel = "Variance expliquée (%)")
g_eig.figure.suptitle("Variance expliquée par dimension")

g_eig = sns.pointplot(x = "Dimension", 
                          y = "% variance expliquée",
                          data = eig)
g_eig.figure.suptitle("Variance expliquée par dimension")

plt.show()

```

```python
#table de contingence des 2 variables: Content Rating et Category
TC=pd.crosstab(index=data_cleaned['Category'], columns=data_cleaned['Content Rating'],margins=False)
TC
```

```python
#calcul des totaux en ligne
tot_lig = np.sum(TC.values,axis=1)
print(tot_lig)

print("----------------------------------------")
#calcul des totaux en colonne
tot_col = np.sum(TC.values,axis=0)
print(tot_col)
#il faut que la somme s1=s2
s1=np.sum(tot_lig)
s2=np.sum(tot_col)
print(s1)
print(s2)
# SI S1=S2=9145---> c'est bon
```

```python
#profil ligne
prof_lig = np.apply_along_axis(arr=TC.values,axis=1,func1d=lambda x:x/np.sum(x))
#le profil marginal des lignes
prof_marg_lig = tot_col/np.sum(tot_col)
print(prof_marg_lig)
```

```python
#xi2--> distance entre lignes
distPairesLig = np.zeros(shape=(prof_lig.shape[0],prof_lig.shape[0]))
for i in range(prof_lig.shape[0]-1):
    for j in range(0,prof_lig.shape[0]):
         distPairesLig[i,j] = np.sum((prof_lig[i,:]-prof_lig[j,:])**2/prof_marg_lig)

xi2=pd.DataFrame(distPairesLig,index=TC.index,columns=TC.index)
xi2.head(2)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})#size de figgure de heatemap en dessous 
```

```python
#Les écarts entre les differentes categories 
sns.heatmap(distPairesLig,vmin=0,vmax=np.max(distPairesLig),linewidth=0.1,cmap='Blues',xticklabels=TC.index,
yticklabels=TC.index)
```

```python
# apres l presentation du distance entre chaque categories on a calculer la Distance entre les categories et  l'origin.=distance au profil moy
distoLig = np.apply_along_axis(arr=prof_lig,axis=1,func1d=lambda x:np.sum((x-prof_marg_lig)**2/prof_marg_lig))
print(pd.DataFrame(distoLig,index=TC.index))
```

```python
#distance des catégories et origine + poids + Inertie pour les lignes (Catégories)
poidsLig = tot_lig/np.sum(tot_lig)
inertieLig = distoLig * poidsLig
print(pd.DataFrame(np.transpose([distoLig,poidsLig,inertieLig]),columns=['Disto2','Poids','Inertie'] ,index=TC.index))
```

```python
#Somme des inerties des lignes
# !!! "art & design" a un profil différent(éloignement de l'origine), mais sont mal représentés (poids), leur inertie ne ressort pas vraiment. 
tot_InertieLig = np.sum(inertieLig)
print(tot_InertieLig)
```

```python
#afc a l'aide de la fonction CA
afc = CA(row_labels=TC.index,col_labels=TC.columns)
afc.fit(TC.values)
#information restituée sur les facteurs
#les val propres
print(afc.eig_)
#print(pd.DataFrame(afc.row_coord_,index=TC.index))
```

```python
#Let us calculate the Euclidean distances in the 1st plane.

distPairesLigF1 = np.zeros(shape=(prof_lig.shape[0],prof_lig.shape[0]))
#double boucle
for i in range(prof_lig.shape[0]-1):
    for j in range(0,prof_lig.shape[0]):
        distPairesLigF1[i,j] = np.sum((afc.row_coord_[i,0]-afc.row_coord_[j,0])**2)

#affichage
print(pd.DataFrame(distPairesLigF1,index=TC.index,columns=TC.index))
```

```python
#Column analysis
#profil marginal des Content Rating
prof_marg_col = tot_lig/np.sum(tot_lig)
print(prof_marg_col)
```

```python
#Les profils colonnes
prof_col = np.apply_along_axis(arr=TC.values,axis=0,func1d=lambda x:x/np.sum(x))
```

```python
#distance à l'orgine des colonnes
distoCol = np.apply_along_axis(arr=prof_col,axis=0,func1d=lambda x:np.sum((x-prof_marg_col)**2/prof_marg_col))
#affichage
print(pd.DataFrame(distoCol,index=TC.columns))
```

```python
#poids de chaque colonne
poidsCol = tot_col/np.sum(tot_col)
print(pd.DataFrame(poidsCol,index=TC.columns))
```

```python
#inertie
inertieCol = distoCol*poidsCol
print(pd.DataFrame(inertieCol,index=TC.columns))
```

```python
#statistique du KHI-2
#effectifs totaux
n = np.sum(TC.values)
#tableau sous indépendance
E = np.dot(np.reshape(tot_lig,(33,1)),np.reshape(tot_col,(1,6)))/n
#khi2
KHI2 = np.sum(((TC.values-E)**2)/E)
print(KHI2)
```

```python
#degré de liberté
ddl = (E.shape[0]-1)*(E.shape[1]-1)
print(ddl)
#p-value 
print(1-scipy.stats.chi2.cdf(KHI2,ddl))
```
